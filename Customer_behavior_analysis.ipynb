What factors are influencing customer engagement.

```python
query="""select Distinct(p.ProductName),p.ProductID,cj.Action,cj.Stage,avg(Likes) as average_likes,e.ContentType from engagement_data as e inner join products as p
on p.ProductID=e.ProductID inner join customer_journey as cj on p.ProductID=cj.ProductID 
where Likes>(select avg(Likes) from engagement_data) and Action="Purchase"  group by ProductName,ProductID,Action,Stage ,ContentType order by ProductID;"""
engage=pd.read_sql(query,connection)
engage_df=pd.DataFrame(engage)
engage_df
```

```python
query="""SELECT e.ContentType, COUNT(DISTINCT e.ProductID) AS Unique_Product,AVG(e.Likes) AS Avg_Likes,COUNT(DISTINCT cj.CustomerID) AS Unique_Customers_In_Journey,
    SUM(CASE WHEN cj.Stage = 'Checkout' AND cj.Action = 'Drop-off' THEN 1 ELSE 0 END) AS Checkout_Dropoffs,
    SUM(CASE WHEN cj.Stage = 'Checkout' AND cj.Action = 'Purchase' THEN 1 ELSE 0 END) AS Completed_Purchases,
    ROUND(AVG(cr.Rating), 2) AS Avg_Product_Rating FROM engagement_data as e LEFT JOIN customer_journey cj ON e.ProductID = cj.ProductID
LEFT JOIN customer_reviews cr ON cj.ProductID = cr.ProductID  GROUP BY e.ContentType ORDER BY Unique_Product DESC;"""
engage_2=pd.read_sql(query,connection)
engage_2_df=pd.DataFrame(engage_2)
engage_2_df
```

```python
from textblob import TextBlob

query="""SELECT * FROM customer_behavior_analysis.customer_reviews;"""
sentiment=pd.read_sql(query,connection)
sentiment_df=pd.DataFrame(sentiment)

reviews=["Average  experience,  nothing  special.",
         "Good  quality,  but  could  be  cheaper",
         "The  quality  is    top-notch.",
         "Five  stars  for  the  quick  delivery.",
         "Product  did    not  meet  my  expectations.",
         "Not  worth  the  money.",
         "The  product  stopped  working  after  a  month.",
         "I  had  a  bad  experience  with  this  product.",
         "The  product    is  okay,  but  the  instructions  were  unclear.",
         "Disappointed  with  the  performance.",
         "The  product  arrived    late.",
         "I  love  this    product,  will  buy  again!",
         "Five  stars  for  the  quick  delivery.",
         "Customer  support  was  very  helpful.",
         "Excellent  product,  highly  recommend!",
         "Great  purchase,  very  satisfied.",
         "Amazing  value  for  the  price.",
         "Shipping  was  fast  and  the  item  was  well-packaged",
         "Exceeded  my  expectations!"]

create_df=pd.DataFrame(reviews)
create_df

def analyze_sentiment(text):
    polarity=TextBlob(text).sentiment.polarity
    if polarity>0:
        return "Positive"
    elif polarity<0:
        return "Negative"
    else:
        return "Neutral" 
        
        
sentiment_df["sentiments"]=sentiment_df["ReviewText"].apply(analyze_sentiment)

sentiment_df
```

```python
sentiment_df.sentiments.value_counts()
```

Identify key complaints and improvemnts area

```python
Negative=sentiment_df[sentiment_df["Rating"]<3]
Negative
```

Find Pattrens between negative reviews and Product performance

```python
query="""with cust_mer as(select  (p.ProductID),p.ProductName,cr.Rating,cr.Reviewtext from customer_reviews as cr inner join 
products as p on p.ProductID=cr.ProductID)
select ProductID,ProductName,Rating,
case ReviewText
when "Average  experience,  nothing  special." then "Product Performance"
when "Product  did    not  meet  my  expectations ." then "Product Performance"
when "Not  worth  the  money." then "Product Performance"
when "The  product  stopped  working  after  a  month." then "Product Performance"
when "I  had  a  bad  experience  with  this  product." then "Product Performance"
when "Disappointed  with  the  performance." then "Product Performance"
else "Others" 
end as product_performance_status from cust_mer order by ProductID;"""
product_per=pd.read_sql(query,connection)
product_per_df=pd.DataFrame(product_per)
product_per_df
```

```python
product_per_df.to_csv("Product_performance.csv",index=False)
```
